```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
```

```python
# a. plot
plt.plot(x, y, marker='o')

# b. customization
plt.title("Simple Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.grid(True)

plt.show()
```

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]
```

```python
# 2a.  plot
plt.bar(students, marks, color=['blue', 'green', 'orange', 'red'])

# 2b. title
plt.title("Students Marks")

plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```

```python
# 3
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]
```

```python
# highlight highest (first one)
explode = [0.1, 0, 0, 0]

plt.pie(revenue, labels=regions, autopct='%1.1f%%', explode=explode)

plt.title("Revenue Distribution by Region")

plt.show()
```

```python
# random data
data = np.random.randint(1, 101, 1000)

# histogram
plt.hist(data, bins=20, color='purple', edgecolor='black')

plt.title("Histogram of Random Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```
