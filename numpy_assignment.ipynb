```python
import numpy as np
```

```python
arr = np.arange(1, 21)

# a. calculations
print("Array:", arr)
print("Sum:", np.sum(arr))
print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))
```

```python
# b. indices > 10
indices = np.where(arr > 10)
print("Indices where elements > 10:", indices)
```

```python
arr2 = np.arange(1, 17).reshape(4, 4)

# a. print
print("2D Array:\n", arr2)
```

```python
# b. transpose
print("Transpose:\n", arr2.T)
```

```python
# c. row & column sum
print("Row-wise sum:", np.sum(arr2, axis=1))
print("Column-wise sum:", np.sum(arr2, axis=0))
```

```python
a = np.random.randint(1, 21, (3, 3))
b = np.random.randint(1, 21, (3, 3))

print("Array A:\n", a)
print("Array B:\n", b)
```

```python
# a. operations
print("Addition:\n", a + b)
print("Subtraction:\n", a - b)
print("Multiplication:\n", a * b)
```

```python
# b. dot product
print("Dot Product:\n", np.dot(a, b))
```

```python
arr3 = np.arange(1, 13)
```

```python
reshaped = arr3.reshape(3, 4)
print("Reshaped Array:\n", reshaped)
```

```python
# slicing
result = reshaped[:2, -2:]
print("Sliced (first 2 rows, last 2 columns):\n", result)
```
